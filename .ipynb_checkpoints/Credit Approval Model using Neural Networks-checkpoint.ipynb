# Impute missing values

# 1. First we are going to deal with numberical values and impute them using mean 

```python
# Impute the missing values with mean imputation
credit.fillna(credit.mean(), inplace=True)

# Count the number of NaNs in the dataset to verify
credit.isnull().sum()
```

# 2. We are going to impute these missing values with the most frequent values as present in the respective columns

```python
# Iterate over each column of cresit
for col in credit.columns:
    # Check if the column is of object type
    if credit[col].dtypes == 'object':
        # Impute with the most frequent value
        credit = credit.fillna(credit[col].value_counts().index[0])

# Count the number of NaNs in the dataset and print the counts to verify
credit.isnull().values.sum()
```

# Feature selection and label encoding

```python
encoded=pd.get_dummies(credit[['A1', 'A4','A5','A6','A7','A9','A10','A12','A13','A16']], prefix=['A1', 'A4','A5','A6','A7','A9','A10','A12','A13','A16'],drop_first=True)
encoded.head()
```

```python
#Let's concatenate the encoded dataframe with the original dataframe-right
credit_updated = pd.concat([credit.drop(['A1', 'A4','A5','A6','A7','A9','A10','A12','A13','A16'],axis=1), encoded],axis=1)
credit_updated
```

# Correlation of features with target variable

```python
credit_updated.corr()["A16_-"].sort_values().plot(kind='bar',figsize=(18,6),color='blue')
plt.title("Feature correlation",color='red',size=14)
plt.ylabel("Coefficient",color='red',size=14)
plt.xlabel("Features",color='red',size=14)
plt.plot()
```

```python
X = credit_updated.drop('A16_-',axis=1)
y = credit_updated.iloc[:,-1]
```

# Feature selection using LASSO

```python
# # LASSO
pipeline = Pipeline([('model',Lasso())])
lambda_ = np.logspace(-2,1)
search = GridSearchCV(pipeline,{'model__alpha':lambda_},
cv = 10, scoring="accuracy")
search.fit(X,y)
search.best_params_
coefficients = search.best_estimator_.named_steps['model'].coef_
coefficients
importance_features = np.abs(coefficients)
# to_positive_Coefficients


selected_features = [i for i in np.array(X.columns)[importance_features > 0]]
selected_features
```

# When we compare the features chosen using LASSO to the preceding feature correlation graph, we can plainly see that features with high correlation were chosen,  while those with poor correlation were removed.

# Plotting selected features by LASSO

```python
plt.figure(figsize=(12,8))
plt.bar(selected_features,np.array(importance_features)[importance_features >0])
plt.xticks(rotation=45)
plt.title("Weights of features with LASSO")
plt.show()
```

# Building a Neural network based model

```python
X= X[selected_features]
#Cross Validation
#Number of K Folds
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
#A Multi Layer Perceptron Classifier

#Fitting the model
NN_model = MLPClassifier(random_state=42)
NN_model.fit(X,y)



mlp_CV_Scores = [np.mean(cross_val_score(NN_model, X, y, cv=cv, scoring="precision")),
 np.mean(cross_val_score(NN_model, X, y, cv=cv, scoring="recall"))]

#print("The accuracy score is: {} ".format(metrics.accuracy_score(y_test, y_pred)))

print("F1:",mlp_CV_Scores[0])
print("recall:",mlp_CV_Scores[1])
```

# Using hyperparameter tuning

# Search space include activation, batch size, learning rate among others

```python
hyperparams={'Multi layer Perceptron':
                 {'model':MLPClassifier(),
                  'params':{
                    'hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
                      'random_state':[42],
                    'activation':['tanh', 'relu','identity', 'logistic'],
                      'batch_size':[400],
                   'solver':['sgd', 'adam','lbfgs'],
                     'alpha': [0.0001, 0.05],
                      'learning_rate':['constant','adaptive','invscaling']
                    }
                 }
            }
```

```python
accuracy_values=[]

for key in hyperparams.keys():
    g_search=GridSearchCV(hyperparams[key]['model'],hyperparams[key]['params'],
                          cv=10,return_train_score=False,verbose=2)
    g_search.fit(X,y)

    accuracy_values.append({'model':key,
                            'best_score':g_search.best_score_,
                            'best_params':g_search.best_params_
                           })
    #loop+=1
```

```python
accuracy_values
```

```python
pd.set_option('display.max_colwidth', None) #using this to avoid strings being truncated in column output
accuracy_df=pd.DataFrame(accuracy_values)

accuracy_df
```
